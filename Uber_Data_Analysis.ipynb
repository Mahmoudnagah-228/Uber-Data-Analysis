```python
import datetime

def get_time_group(time_obj):
    if time_obj >= datetime.time(7, 0, 0) and time_obj < datetime.time(9, 0, 0):
        return 'early morning'
    elif time_obj >= datetime.time(9, 0, 0) and time_obj < datetime.time(12, 0, 0):
        return 'morning'
    elif time_obj >= datetime.time(12, 0, 0) and time_obj < datetime.time(15, 0, 0):
        return 'noon'
    elif time_obj >= datetime.time(15, 0, 0) and time_obj < datetime.time(18, 0, 0):
        return 'afternoon'
    elif time_obj >= datetime.time(18, 0, 0) and time_obj < datetime.time(21, 0, 0):
        return 'evening'
    elif time_obj >= datetime.time(21, 0, 0) or time_obj < datetime.time(0, 0, 0):
        return 'midnight'
    elif time_obj >= datetime.time(0, 0, 0) and time_obj < datetime.time(7, 0, 0):
        return 'dancing'
    else:
        return 'unknown'

data['time_group'] = data['time_of_day'].apply(get_time_group)

display(data[['time_of_day', 'time_group']].head())
```
